```python
def bottom_up_fibonacci(N):
    if N == 0 or N == 1:
        return N

    A = 0
    B = 1

    count = 0  

    for _ in range(2, N + 1):
        Temp = A + B
        A = B
        B = Temp
        count += 1  

    return count

counts = [bottom_up_fibonacci(i) for i in range(5, 51)]

import matplotlib.pyplot as plt

plt.plot(range(5, 51), counts)
plt.xlabel('N')
plt.ylabel('Number of Fibonacci(4) Calculations')
plt.title('Number of Fibonacci(4) Calculations for Fibonacci(5) to Fibonacci(50)')
plt.show()
```
